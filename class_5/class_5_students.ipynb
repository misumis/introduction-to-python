```python
TEST_STRING.replace("Is", "was")
```

```python
TEST_STRING.strip()
```

## Time Series Analysis

Time series analysis is a statistical technique that deals with time series data, or trend analysis. Time series data means that data is in a series of particular time periods or intervals. Typical aproaches to time series analysis include autoregressive models, moving average models, and state space models.

In our case we are going to use the `ARIMA` model. ARIMA stands for AutoRegressive Integrated Moving Average. ARIMA models are denoted with the notation ARIMA(p, d, q). These three parameters account for seasonality, trend, and noise in data.

ARIMA Models are specified by three order parameters: (p, d, q), where,
- p is the order of the AR term
- q is the order of the MA term
- d is the number of differencing required to make the time series stationary

### Types of ARIMA Models

- ARIMA : Non-seasonal Autoregressive Integrated Moving Averages
  - ARIMA(p,0,0) = AR(p)
  
  - ARIMA(0,d,0) = I(d)
  - ARIMA(0,0,q) = MA(q)
  - ARIMA(p,d,q) = AR(p) + I(d) + MA(q)
  

- SARIMA : Seasonal ARIMA
  - SARIMA(p,d,q)(P,D,Q)m = ARIMA(p,d,q) × ARIMA(P,D,Q)m

- SARIMAX : Seasonal ARIMA with Exogenous Variable
  
  - SARIMAX(p,d,q)(P,D,Q)m = ARIMA(p,d,q) × ARIMA(P,D,Q)m × Xt



SARIMA : Seasonal ARIMA
SARIMAX : Seasonal ARIMA with exogenous variables

To apply ARIMA model, first we need to check if the data is stationary. We can use the Augmented Dickey-Fuller test to check if the data is stationary.

```python
from statsmodels.tsa.stattools import adfuller
from numpy import log
result = adfuller(df.value.dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

The null hypothesis of the Augmented Dickey-Fuller test is that there is a unit root, with the alternative that there is no unit root. If the p-value is above a critical size, then we cannot reject that there is a unit root. If the p-value is below a critical size, then we reject that there is a unit root.

In our case the p-value is 1.00 which means that we cannot reject that there is a unit root. Therefore, we need to make the data stationary.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})


# Original Series
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(df.value); axes[0, 0].set_title('Original Series')
plot_acf(df.value, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(df.value.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df.value.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(df.value.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df.value.diff().diff().dropna(), ax=axes[2, 1])

plt.show()
```

We can observe that the data reaches a stationary state after the second differencing.

### Finding AR and MA terms (p and q)

To find the number of AR and MA terms, we can use the ACF (AutoCorrelation Function) and PACF (Partial AutoCorrelation Function) plots.

The ACF plot shows the correlation of the series with itself, lagged by x time units. For example at lag 5, ACF would compare series at time instant ‘t1’…’tn’ with series at instant ‘t1-5’…’tn-5’ (t1-5 and tn being end points).

The PACF plot is a plot of the partial correlation coefficients between the series and lags of itself. A partial autocorrelation is a summary of the relationship between an observation in a time series with observations at prior time steps with the relationships of intervening observations removed.

```python
# PACF plot of 1st differenced series to determine AR term
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(df.value.diff()); axes[0].set_title('1st Differencing')
axes[1].set(ylim=(0,5))
plot_pacf(df.value.diff().dropna(), ax=axes[1])

plt.show()
```

We can see that the PACF lag 1 is quite significant since it is well above the significance line. So, we will fix the value of p as 1.

```python
# ACF plot of 1st differenced series to determine MA term
from statsmodels.graphics.tsaplots import plot_acf
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(df.value.diff()); axes[0].set_title('1st Differencing')
axes[1].set(ylim=(0,1.2))
plot_acf(df.value.diff().dropna(), ax=axes[1])

plt.show()
```

We can observe 2 significant spikes in the ACF plot at lags 1 and 2. Therefore, we can assume that the value of q is 2.

### ARIMA model

```python
from statsmodels.tsa.arima.model import ARIMA

# 1,1,2 ARIMA Model
model = ARIMA(endog=df.value, order=(1,1,2))
model_fit = model.fit()
model_fit.summary()
```

### Forecasting

We can now use the ARIMA model to forecast the future values of the time series.

First we need to divide the data into train and test sets. We will use the first 70% of the data as the training set and the remaining 30% as the test set.

```python
# Create Training and Test

division_index = round(len(df)*0.7)
items_left = len(df) - division_index

(train, test) = df.value[:division_index], df.value[division_index:]
```

```python
# Build Model
# model = ARIMA(train, order=(3,2,1))  
model = ARIMA(train, order=(1, 1, 1), trend='t')  
fitted = model.fit()  

# Forecast
fc_df = fitted.get_forecast(items_left).summary_frame()

# Make as pandas series
fc_series = pd.Series(fc_df['mean'], index=test.index)
lower_series = pd.Series(fc_df['mean_ci_lower'], index=test.index)
upper_series = pd.Series(fc_df['mean_ci_upper'], index=test.index)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.15)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

### Measuring accuracy of the model

Most common accuracy metrics for time series forecasting are 
- MAPE (Mean Absolute Percentage Error), 
- ME (Mean Error),
- MAE (Mean Absolute Error), 
- MPE (Mean Percentage Error), 
- RMSE (Root Mean Squared Error), 
- ACF1 (Autocorrelation at lag 1),
- CORR (Correlation between the actual and forecasted values).
- MINMAX (Min-Max Error),

```python
from statsmodels.tsa.stattools import acf
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    acf1 = acf(forecast-actual)[1]                      # ACF1
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse, 'acf1':acf1, 
            'corr':corr, 'minmax':minmax})

forecast_accuracy(fc_df['mean'], test.values)
```

MAPE on a level 16.99% is quite good. It implies that the model is about 83% accurate in predicting the next 61 observations.
