```python
# create a function that takes an initial hour and adds 15 minutes to it and returns result in a format HH:MM - HH:MM
# example: 10:00 - 10:15
import datetime

def add_15_minutes(staring_hour):
    start = datetime.datetime.strptime(staring_hour, '%H:%M')
    end = start + datetime.timedelta(minutes=15)
    result = f"{start.strftime('%H:%M')} - {end.strftime('%H:%M')}"
    return result
  
add_15_minutes('12:15')

# create a function that takes a starting hour and ending hour and returns 15 minutes time slots between them in a format HH:MM - HH:MM
# example: 10:00 - 10:15
import datetime
 
def get_15_minutes_time_slots(starting_hour, ending_hour):
    start = datetime.datetime.strptime(starting_hour, '%H:%M')
    end = datetime.datetime.strptime(ending_hour, '%H:%M')
    result = []
    while start < end:
        result.append(f"{start.strftime('%H:%M')} - {(start + datetime.timedelta(minutes=12)).strftime('%H:%M')}")
        start = start + datetime.timedelta(minutes=12)
    return result
  
pd.DataFrame(get_15_minutes_time_slots('12:00', '15:35'))
```
